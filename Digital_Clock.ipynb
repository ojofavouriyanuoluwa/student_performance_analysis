```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Digital Clock Dashboard with Multiple Features
A comprehensive time management tool for Jupyter Notebook
"""

import ipywidgets as widgets
from IPython.display import display, clear_output
import datetime
import time
from threading import Thread
import pytz

# ============= DIGITAL CLOCK TAB =============
def create_digital_clock():
    """Creates a simple, elegant digital clock with date display"""
    
    # Create display widgets
    time_label = widgets.HTML(
        value="<h1 style='text-align:center; font-size:64px; font-family:monospace; color:#2c3e50;'>--:--:--</h1>"
    )
    date_label = widgets.HTML(
        value="<h3 style='text-align:center; font-family:monospace; color:#7f8c8d;'>--</h3>"
    )
    
    # Style the container
    container = widgets.VBox([
        widgets.HTML("<h2 style='text-align:center; color:#3498db;'>🕐 Digital Clock</h2>"),
        time_label,
        date_label,
        widgets.HTML("<hr>"),
        widgets.HTML("<p style='text-align:center; color:#95a5a6; font-size:12px;'>Live 24-hour format</p>")
    ], layout=widgets.Layout(
        padding='20px',
        border='2px solid #3498db',
        border_radius='10px',
        margin='10px'
    ))
    
    def update_clock():
        """Update the clock every second"""
        while True:
            now = datetime.datetime.now()
            time_str = now.strftime("%H:%M:%S")
            date_str = now.strftime("%A, %B %d, %Y")
            
            time_label.value = f"<h1 style='text-align:center; font-size:64px; font-family:monospace; color:#2c3e50;'>{time_str}</h1>"
            date_label.value = f"<h3 style='text-align:center; font-family:monospace; color:#7f8c8d;'>{date_str}</h3>"
            time.sleep(1)
    
    # Start the update thread
    thread = Thread(target=update_clock, daemon=True)
    thread.start()
    
    return container

# ============= INTERACTIVE CLOCK TAB =============
def create_interactive_clock():
    """Creates a customizable clock with user controls"""
    
    # Time display area
    time_display = widgets.HTML(
        value="<h1 style='text-align:center; font-size:56px; font-family:monospace; color:#e67e22;'>--:--:--</h1>"
    )
    
    # Control widgets
    format_selector = widgets.RadioButtons(
        options=['24-hour format', '12-hour format'],
        value='24-hour format',
        description='Time format:',
        style={'description_width': 'initial'}
    )
    
    seconds_toggle = widgets.Checkbox(
        value=True,
        description='Show seconds',
        style={'description_width': 'initial'}
    )
    
    date_toggle = widgets.Checkbox(
        value=True,
        description='Show date',
        style={'description_width': 'initial'}
    )
    
    color_picker = widgets.ColorPicker(
        value='#e67e22',
        description='Clock color:',
        style={'description_width': 'initial'}
    )
    
    # Control panel
    controls = widgets.VBox([
        widgets.HTML("<h3>🎮 Display Controls</h3>"),
        format_selector,
        seconds_toggle,
        date_toggle,
        color_picker
    ], layout=widgets.Layout(
        padding='10px',
        border='1px solid #bdc3c7',
        border_radius='5px',
        margin='5px'
    ))
    
    # Container
    container = widgets.VBox([
        widgets.HTML("<h2 style='text-align:center; color:#e67e22;'>🎮 Interactive Clock</h2>"),
        controls,
        time_display
    ])
    
    def update_interactive_clock():
        """Update clock based on user preferences"""
        while True:
            now = datetime.datetime.now()
            
            # Format time based on user choice
            if format_selector.value == '24-hour format':
                if seconds_toggle.value:
                    time_str = now.strftime("%H:%M:%S")
                else:
                    time_str = now.strftime("%H:%M")
            else:  # 12-hour format
                if seconds_toggle.value:
                    time_str = now.strftime("%I:%M:%S %p").lstrip('0')
                else:
                    time_str = now.strftime("%I:%M %p").lstrip('0')
            
            # Build display text
            display_text = time_str
            if date_toggle.value:
                date_str = now.strftime("%A, %B %d, %Y")
                display_text += f"<br><span style='font-size:24px;'>{date_str}</span>"
            
            # Update display with chosen color
            time_display.value = f"<h1 style='text-align:center; font-size:56px; font-family:monospace; color:{color_picker.value};'>{display_text}</h1>"
            
            time.sleep(1)
    
    thread = Thread(target=update_interactive_clock, daemon=True)
    thread.start()
    
    return container

# ============= ALARM CLOCK TAB =============
def create_alarm_clock():
    """Creates a functional alarm clock with visual alerts"""
    
    # Current time display
    current_time = widgets.HTML(
        value="<h1 style='text-align:center; font-size:56px; font-family:monospace; color:#e74c3c;'>--:--:--</h1>"
    )
    
    # Alarm settings
    alarm_hour = widgets.IntSlider(
        min=0, max=23, value=7, 
        description='Hour (0-23):',
        style={'description_width': 'initial'},
        layout=widgets.Layout(width='300px')
    )
    
    alarm_minute = widgets.IntSlider(
        min=0, max=59, value=0, 
        description='Minute (0-59):',
        style={'description_width': 'initial'},
        layout=widgets.Layout(width='300px')
    )
    
    # Buttons
    set_alarm_btn = widgets.Button(
        description='🔔 Set Alarm', 
        button_style='warning',
        layout=widgets.Layout(width='120px')
    )
    
    clear_alarm_btn = widgets.Button(
        description='❌ Clear Alarm', 
        button_style='danger',
        layout=widgets.Layout(width='120px')
    )
    
    # Status display
    alarm_status = widgets.HTML(value="<p style='color: #7f8c8d;'>No alarm set</p>")
    
    # Alarm state
    alarm_time = None
    alarm_active = False
    alarm_triggered = False
    
    def set_alarm(b):
        nonlocal alarm_time, alarm_active, alarm_triggered
        alarm_time = f"{alarm_hour.value:02d}:{alarm_minute.value:02d}"
        alarm_active = True
        alarm_triggered = False
        alarm_status.value = f"<p style='color: #27ae60; font-weight: bold;'>✓ Alarm set for {alarm_time}</p>"
    
    def clear_alarm(b):
        nonlocal alarm_active, alarm_triggered
        alarm_active = False
        alarm_triggered = False
        alarm_status.value = "<p style='color: #7f8c8d;'>No alarm set</p>"
    
    set_alarm_btn.on_click(set_alarm)
    clear_alarm_btn.on_click(clear_alarm)
    
    def update_alarm_clock():
        nonlocal alarm_triggered
        while True:
            now = datetime.datetime.now()
            time_str = now.strftime("%H:%M:%S")
            current_time.value = f"<h1 style='text-align:center; font-size:56px; font-family:monospace; color:#e74c3c;'>{time_str}</h1>"
            
            # Check if alarm should trigger
            if alarm_active and not alarm_triggered:
                current_time_str = now.strftime("%H:%M")
                if current_time_str == alarm_time:
                    alarm_triggered = True
                    # Trigger alarm with visual effects
                    for _ in range(10):
                        current_time.value = "<h1 style='text-align:center; font-size:56px; font-family:monospace; color:#ff0000;'>🔔 ALARM! 🔔</h1>"
                        alarm_status.value = "<p style='color: #ff0000; font-weight: bold; font-size: 16px;'>🔔🔔🔔 WAKE UP! TIME TO RISE! 🔔🔔🔔</p>"
                        time.sleep(0.5)
                        current_time.value = f"<h1 style='text-align:center; font-size:56px; font-family:monospace; color:#e74c3c;'>{time_str}</h1>"
                        time.sleep(0.5)
                    
                    alarm_status.value = "<p style='color: #ff0000; font-weight: bold;'>⏰ Alarm has been triggered!</p>"
            
            time.sleep(1)
    
    thread = Thread(target=update_alarm_clock, daemon=True)
    thread.start()
    
    # Layout
    alarm_controls = widgets.VBox([
        widgets.HTML("<h3>⏰ Alarm Settings</h3>"),
        alarm_hour,
        alarm_minute,
        widgets.HBox([set_alarm_btn, clear_alarm_btn]),
        alarm_status
    ], layout=widgets.Layout(
        padding='15px',
        border='1px solid #e74c3c',
        border_radius='5px',
        margin='10px'
    ))
    
    return widgets.VBox([
        widgets.HTML("<h2 style='text-align:center; color:#e74c3c;'>⏰ Alarm Clock</h2>"),
        current_time,
        alarm_controls
    ])

# ============= WORLD CLOCK TAB =============
def create_world_clock():
    """Displays time in multiple time zones"""
    
    # Define time zones
    timezones = {
        '🌍 UTC': 'UTC',
        '🇺🇸 New York': 'America/New_York',
        '🇬🇧 London': 'Europe/London',
        '🇯🇵 Tokyo': 'Asia/Tokyo',
        '🇦🇪 Dubai': 'Asia/Dubai',
        '🇦🇺 Sydney': 'Australia/Sydney',
        '🇩🇪 Berlin': 'Europe/Berlin',
        '🇮🇳 Mumbai': 'Asia/Kolkata',
        '🇧🇷 Sao Paulo': 'America/Sao_Paulo',
        '🇿🇦 Johannesburg': 'Africa/Johannesburg'
    }
    
    # Create display areas
    clock_displays = {}
    
    # Create a grid of clocks
    clock_grid = widgets.GridBox(
        layout=widgets.Layout(
            grid_template_columns="repeat(2, 1fr)",
            grid_gap="10px",
            padding="10px"
        )
    )
    
    # Create HTML widgets for each timezone
    for city, tz in timezones.items():
        display_box = widgets.HTML(
            value=f"""
            <div style='border:1px solid #3498db; border-radius:5px; padding:10px; text-align:center;'>
                <strong>{city}</strong><br>
                <span style='font-size:20px; font-family:monospace;'>--:--:--</span>
            </div>
            """
        )
        clock_displays[tz] = display_box
        clock_grid.children += (display_box,)
    
    def update_world_clocks():
        """Update all world clocks simultaneously"""
        while True:
            for tz_name, display in clock_displays.items():
                try:
                    tz = pytz.timezone(tz_name)
                    current_time = datetime.datetime.now(tz)
                    time_str = current_time.strftime("%H:%M:%S")
                    
                    # Update the display
                    city_name = [city for city, tz in timezones.items() if tz == tz_name][0]
                    display.value = f"""
                    <div style='border:1px solid #3498db; border-radius:5px; padding:10px; text-align:center; background:#f8f9fa;'>
                        <strong>{city_name}</strong><br>
                        <span style='font-size:20px; font-family:monospace; color:#2c3e50;'>{time_str}</span>
                    </div>
                    """
                except Exception as e:
                    display.value = f"<div style='color:red;'>Error: {tz_name}</div>"
            
            time.sleep(1)
    
    thread = Thread(target=update_world_clocks, daemon=True)
    thread.start()
    
    return widgets.VBox([
        widgets.HTML("<h2 style='text-align:center; color:#3498db;'>🌍 World Clock</h2>"),
        widgets.HTML("<p style='text-align:center; color:#7f8c8d;'>Real-time times across major cities</p>"),
        clock_grid
    ])

# ============= STOPWATCH TAB =============
def create_stopwatch():
    """Creates a functional stopwatch with start/stop/reset controls"""
    
    # Display
    time_display = widgets.HTML(
        value="<h1 style='text-align:center; font-size:64px; font-family:monospace; color:#27ae60;'>00:00:00</h1>"
    )
    
    # Buttons
    start_btn = widgets.Button(description='▶ Start', button_style='success', layout=widgets.Layout(width='100px'))
    stop_btn = widgets.Button(description='⏸ Stop', button_style='warning', layout=widgets.Layout(width='100px'))
    reset_btn = widgets.Button(description='🔄 Reset', button_style='danger', layout=widgets.Layout(width='100px'))
    
    # State variables
    running = False
    elapsed_seconds = 0
    last_update = 0
    
    def update_stopwatch():
        nonlocal elapsed_seconds, last_update, running
        while True:
            if running:
                elapsed_seconds += 1
                hours = elapsed_seconds // 3600
                minutes = (elapsed_seconds % 3600) // 60
                seconds = elapsed_seconds % 60
                time_str = f"{hours:02d}:{minutes:02d}:{seconds:02d}"
                time_display.value = f"<h1 style='text-align:center; font-size:64px; font-family:monospace; color:#27ae60;'>{time_str}</h1>"
            time.sleep(1)
    
    def start_stopwatch(b):
        nonlocal running
        running = True
    
    def stop_stopwatch(b):
        nonlocal running
        running = False
    
    def reset_stopwatch(b):
        nonlocal elapsed_seconds, running
        running = False
        elapsed_seconds = 0
        time_display.value = "<h1 style='text-align:center; font-size:64px; font-family:monospace; color:#27ae60;'>00:00:00</h1>"
    
    start_btn.on_click(start_stopwatch)
    stop_btn.on_click(stop_stopwatch)
    reset_btn.on_click(reset_stopwatch)
    
    button_row = widgets.HBox([start_btn, stop_btn, reset_btn], layout=widgets.Layout(justify_content='center'))
    
    thread = Thread(target=update_stopwatch, daemon=True)
    thread.start()
    
    return widgets.VBox([
        widgets.HTML("<h2 style='text-align:center; color:#27ae60;'>⏱️ Stopwatch</h2>"),
        time_display,
        button_row,
        widgets.HTML("<p style='text-align:center; color:#95a5a6; font-size:12px;'>Click Start to begin timing</p>")
    ], layout=widgets.Layout(padding='20px'))

# ============= MAIN DASHBOARD =============
def create_dashboard():
    """Create the complete dashboard with all tabs"""
    
    # Create all clock widgets
    digital_clock = create_digital_clock()
    interactive_clock = create_interactive_clock()
    alarm_clock = create_alarm_clock()
    world_clock = create_world_clock()
    stopwatch = create_stopwatch()
    
    # Create tab widget
    tabs = widgets.Tab()
    tabs.children = [
        digital_clock,
        interactive_clock,
        alarm_clock,
        world_clock,
        stopwatch
    ]
    
    # Set tab titles
    tabs.set_title(0, '🕐 Digital Clock')
    tabs.set_title(1, '🎮 Interactive Clock')
    tabs.set_title(2, '⏰ Alarm Clock')
    tabs.set_title(3, '🌍 World Clock')
    tabs.set_title(4, '⏱️ Stopwatch')
    
    # Dashboard header with instructions
    header = widgets.HTML("""
    <div style="text-align: center; padding: 20px; 
                background: linear-gradient(135deg, #667eea 0%, #764ba2 100%);
                color: white; border-radius: 10px; margin-bottom: 20px;">
        <h1>🕐 Digital Clock Dashboard</h1>
        <p>Complete time management solution with multiple features</p>
        <div style="background: rgba(255,255,255,0.2); padding: 10px; border-radius: 5px; margin-top: 10px;">
            <p style="margin: 0;">✅ <strong>All features are fully functional!</strong></p>
            <p style="margin: 5px 0 0 0; font-size: 12px;">💡 Click on different tabs to explore each feature</p>
        </div>
    </div>
    """)
    
    dashboard = widgets.VBox([header, tabs])
    return dashboard

# ============= RUN THE DASHBOARD =============
if __name__ == "__main__" or True:  # This ensures it runs in Jupyter
    print("="*60)
    print("🕐 DIGITAL CLOCK DASHBOARD")
    print("="*60)
    print("\n📋 FEATURES INCLUDED:")
    print("  ✓ Tab 1 - Digital Clock (Large display with date)")
    print("  ✓ Tab 2 - Interactive Clock (Customizable format & colors)")
    print("  ✓ Tab 3 - Alarm Clock (Set alarms with visual alerts)")
    print("  ✓ Tab 4 - World Clock (Multiple time zones)")
    print("  ✓ Tab 5 - Stopwatch (Start/Stop/Reset timer)")
    print("\n🚀 Loading dashboard...")
    print("="*60 + "\n")
    
    # Create and display the dashboard
    dashboard = create_dashboard()
    display(dashboard)
    
    print("\n✅ Dashboard successfully loaded!")
    print("💡 TIP: If any tab appears empty, click on it to activate")
```
